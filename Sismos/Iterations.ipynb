```python
import sys 
sys.path.append('..')

from procesamiento import generarDataset
from framework import Framework
import pandas as pd
```

```python
for inputs in range(10,40):
    print("#####################################################################3")
    print("Inputs",inputs)
    dataset = "Datasets/dataset50_iter_entradas_{}.csv".format(inputs)
    print("Dataset",dataset)
    generarDataset(
        nombre="Enlaces/lista-temporal-50.csv",
        nombreSalida=dataset,
        offset = inputs+1
    )
    df = pd.read_csv(dataset)
    fw = Framework(
        dataframe  = df,
        inputs     = ['{}'.format(i) for i in range(inputs)],
        target     = str(inputs),
        save_dir   = "Sismos\Modelos",
        test_size  = 0.15,
        random_state=0
    )
    #1000,900,800,700,600,500,400,300,200,
    hidden_layers = (inputs,500,400,300,200,100,50)
    max_iter = 100
    print('Layers,Iterations')
    print(hidden_layers,max_iter)
    # print("Processing SVM")
    # fw.get_svm_model()
    # print("SVM RMSE:",fw.get_svm_rmse())
    # print("Saving SVM")
    # fw.save_svm_model('svm_distancia_50_itr_entradas_{}'.format(inputs))
    # print("Processing MLP")
    # fw.get_mlp_model(hidden_layers=hidden_layers,max_iter=max_iter)
    # print("MLP RMSE:",fw.get_mlp_rmse())
    # print("Saving MLP")
    # fw.save_mlp_model('mlp_distancia_50_itr_entradas_{}'.format(inputs))
    print("Processing Keras")
    fw.get_sequential_model(hidden_layers=hidden_layers, epochs=max_iter)
    print("Sequential RMSE:",fw.get_sequential_rmse())
    print("Saving Sequential")
    fw.save_sequential_model('sequential_distancia_50_itr_entradas_{}'.format(inputs))
```
